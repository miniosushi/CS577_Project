```python
import pandas as pd
import numpy as np

file_path = '../Data/Air_Quality.csv' 
df = pd.read_csv(file_path)

df.head()
```

```python
# data types and missing values
df.info()

# missing values in each column
df.isnull().sum()
```

```python
df.describe()
```

```python
# df['Message'].fillna('No Message', inplace=True)
```

```python
df['Start_Date'] = pd.to_datetime(df['Start_Date'], errors='coerce')
# make start date standardized format easy to use 
```

```python
df.head()
```

```python
# Check for duplicates
duplicates = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")

# Drop duplicates
df.drop_duplicates(inplace=True)
```

```python
# List unique values in 'Indicator ID' column
unique_indicator_ids = df['Indicator ID'].unique()
print("Unique values in 'Indicator ID':")
print(unique_indicator_ids)

# List unique values in 'Name' column
unique_names = df['Name'].unique()
print("\nUnique values in 'Name':")
print(unique_names)
```

```python
# Filter to include only NO2 and PM2.5 data
filtered_df = df[df['Name'].isin(['Nitrogen dioxide (NO2)', 'Fine particles (PM 2.5)'])]

# Check the size of the filtered data
print(f"Number of rows in the filtered data: {filtered_df.shape[0]}")
```

```python
# make sure it actually filtered
print("Unique values in 'Name' after filtering:")
print(filtered_df['Name'].unique())

print("\nUnique values in 'Indicator ID' after filtering:")
print(filtered_df['Indicator ID'].unique())
```

```python
# Standardize 'Geo Place Name' values
filtered_df.loc[:, 'Geo Place Name'] = filtered_df['Geo Place Name'].str.lower().str.strip()
```

```python
# Define a function to categorize the 'Time Period'
def categorize_period(period):
    if "Annual Average" in period:
        return "Annual Average"
    elif "Winter" in period or "Summer" in period:
        return "Seasonal"
    elif "-" in period:  # To catch ranges like "2012-2014"
        return "Range"
    else:
        return "Other"

# Apply the function to create the 'Period Type' column
filtered_df.loc[:, 'Period Type'] = filtered_df['Time Period'].apply(categorize_period)
```

```python
# Check the count of each category
print(filtered_df['Period Type'].value_counts())

# Preview the categorized data
filtered_df[['Time Period', 'Period Type']].head(10)
```

```python
filtered_df.head()
```

```python
unique_indicator_geo_place = df['Geo Place Name'].unique()
print("Unique values in 'Geo Place Name':")
print(unique_indicator_geo_place)
```

```python
# Separate DataFrames for annual averages and seasonal data
annual_df = filtered_df[filtered_df['Period Type'] == 'Annual Average']
seasonal_df = filtered_df[filtered_df['Period Type'] == 'Seasonal']

# Save the annual averages data to a CSV file in the Data directory
annual_df.to_csv('../Data/annual_averages_data.csv', index=False)

# Save the seasonal data to a CSV file in the Data directory
seasonal_df.to_csv('../Data/seasonal_data.csv', index=False)
```

```python
file_path = '../Data/Median_Incomes_Neighborhood.csv' 
median_income = pd.read_csv(file_path)

median_income.head()
```

```python
print(median_income['Location'].unique())
```

```python
# Mapping of dataset neighborhood names to your desired names
rename_mapping = {
    'Chelsea/Clinton': 'Chelsea-Village',
    'Battery Park/Tribeca': 'Financial District (CD1)',
    'Greenwich Village': 'Greenwich Village and Soho (CD2)',
    'Washington Heights': 'Washington Heights',
    'Lower East Side': 'Lower East Side and Chinatown (CD3)',
    'Central Harlem': 'Central Harlem (CD10)',
    'Upper East Side': 'Upper East Side (CD8)',
    'Upper West Side': 'Upper West Side',
    'Murray Hill/Stuyvesant': 'Stuyvesant Town and Turtle Bay (CD6)'
}

```

```python
# Filter the dataset for matched neighborhoods
filtered_income = median_income[median_income['Location'].isin(rename_mapping.keys())]

# Rename the neighborhoods to match your desired names
filtered_income.loc[:, 'Location'] = filtered_income['Location'].replace(rename_mapping)

# Verify the changes
print(filtered_income['Location'].unique())
```

```python
filtered_income.to_csv('../Data/Filtered_Neighborhoods.csv', index=False)
```
